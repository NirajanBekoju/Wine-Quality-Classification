```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the datset
df = pd.read_csv("WineQT.csv")

# Overview of dataset
df.head()
```

```python
# Dimension of dataset
df.shape
```

```python
# check data types
df.dtypes
```

```python
# Checking for null values if present
df.isnull().sum()
```

```python
# Descriptive Stats of dataset
df.describe()
```

```python
df.drop('Id', axis = 1 ,  inplace = True)
```

```python
# Check for outliers and range or scale of data on all features
sns.boxplot(data=df, orient="h", palette="Set2")
plt.show()
```

```python
# Check for correlations between features in the dataset
plt.figure(figsize = (10, 8))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# Let's check the distribution of target variable 'quality'
sns.boxplot(x='quality', y='volatile acidity', data=df)
```

```python
sns.boxplot(x='quality', y='alcohol', data=df)
```

```python
sns.violinplot(data=df, x ='quality', y='sulphates')
```
